```python
class Node:
  def __init__(self,prev = None,item = None, next = None):
    self.prev = prev
    self.item = item
    self.next = next
```

```python
class Deque:
  def __init__(self):
    self.front = None
    self.rear = None
    self.count_item = 0

  def is_empty(self):
    if self.count_item == 0:
      print("Deque is empty")
    else:
      print("Deque is not empty")

  def insert_front(self,data):
    n = Node(self.front,data,self.front)
    if self.count_item == 0:
      self.rear = n
    else:
      n.next = self.front.prev
    self.front = n
    self.count_item += 1

  def insert_rear(self,data):
    n = Node(self.rear.next,data,self.rear)
    if self.count_item == 0:
      self.front = n
    else:
      n.prev = self.rear.next
    self.rear = n
    self.count_item += 1

  def total_items(self):
    print("Total item ",self.count_item)

  def get_front(self):
    if self.count_item == 0:
      raise IndexError("Deque is empty")
    else:
      print("Front item is ",self.front.item)

  def get_rear(self):
    if self.count_item == 0:
      raise IndexError("Deque is empty")
    else:
      print("Rear item is ",self.rear.item)

  def delete_rear(self):
    if self.count_item == 0:
      raise IndexError("Deque is empty")
    else:
      self.rear.prev = self.rear
      self.rear  = self.rear.next
      self.count_item -= 1

  def delete_front(self):
    if self.count_item == 0:
      raise IndexError("Deque is empty")
    else:
      self.front.next = self.front
      self.front = self.front.prev
      self.count_item -= 1

```

```python
d1 = Deque()
d1.insert_front(34)
d1.insert_front(35)
d1.insert_rear(36)
d1.insert_rear(37)
d1.total_items()
d1.get_front()
d1.get_rear()
d1.delete_rear()
d1.delete_front()
print()
d1.total_items()
d1.get_front()
d1.get_rear()
```
