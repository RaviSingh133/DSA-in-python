# Queue using List

```python
class Queue:
  def __init__(self):
    self.item = []

  def is_empty(self):
    if len(self.item) == 0:
      print("Queue is empty")
    else:
      print("Queue is not empty")

  def enqueue(self,data):
    self.item.append(data)

  def dequeue(self):
    if not self.is_empty():
      self.item.pop(0)

  def top_value(self):
    print("The top presented value is ",self.item[-1])

  def last_value(self):
    print("The last presented value is ",self.item[0])


  def size(self):
    print("the size of the Queue is ",len(self.item))
```

```python
q1 =Queue()
q1.is_empty()
q1.enqueue(10)
q1.enqueue(20)
q1.enqueue(30)
q1.enqueue(40)
q1.top_value()
q1.size()
q1.enqueue(50)
q1.enqueue(60)
q1.size()
q1.dequeue()
q1.last_value()
q1.size()
q1.last_value()
q1.top_value()
```
