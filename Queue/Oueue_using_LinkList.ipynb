```python
class Node:
  def __init__(self,item = None, next = None):
    self.item = item
    self.next = next
```

```python
class Queue:

  def __init__(self):
    self.front = None
    self.rear = None
    self.count_item = 0

  def is_empty(self):
    if self.count_item == 0:
      print("Queue is empty")
    else:
      print("Queue is not empty")

  def enqueue(self,data):
    n = Node(data)
    if self.count_item == 0:
      self.front = n
    else:
      self.rear.next = n
    self.rear = n
    self.count_item += 1

  def dequeue(self):
    if self.count_item == 0:
      raise IndexError("Queue is empty")
    elif self.front == self.rear:
      self.front = None
      self.rear = None
      self.count_item -= 1
    else:
      self.front = self.front.next
      self.count_item -= 1

  def get_front(self):
    print("The front value is ",self.front.item)

  def get_rear(self):
    print("The rear value is ",self.rear.item)

  def total_item(self):
    print("Total items is ",self.count_item)
```

```python
q1 =Queue()
q1.enqueue(10)
q1.enqueue(20)
q1.enqueue(30)
q1.total_item()
q1.get_front()
q1.get_rear()
q1.dequeue()
print()
q1.total_item()
q1.get_front()
q1.get_rear()
```
