```python
class Node:

  def __init__(self, prev=None, item=None, next=None):
    self.prev = prev
    self.item = item
    self.next = next
```

```python
from ast import Not
class Dll:

  def __init__(self,start= None):
    self.start = start



  def is_empty(self):
    return self.start == None



  def insert_start(self,data):
    n = Node(None,data,self.start)
    if self.start is not None:
      self.start.prev = n
    self.start = n


  def print_list(self):
    temp =self.start
    if temp is not None:
      while temp is not None:
        print(temp.item,end =" ")
        temp = temp.next
      return None
    else:
      print("List is empty")


  def insert_last(self,data):
    temp = self.start
    n= Node(temp,data,None)
    if temp is not None:
      while temp.next is not None:
        temp = temp.next
      temp.next = n
      n.prev = temp
    else:
      self.start = n


  def search_item(self,data):
    temp = self.start
    if temp is not None:
      while temp is not None:
        if temp.item == data:
          print(temp.item," is available")
          break
        temp = temp.next
      else:
        print("value not available")
    else:
      print("list is empty")


  def insert_after(self,temp,data):
    t1 = self.start
    n = Node(t1,data,t1.next)
    if t1 is None:
      pass
    elif t1.next is None:
      if t1.item == temp:
        t1.next = n
        n.prev = t1.next
    else:
      while t1 is not None:
        if t1.item == temp:
          t1.next = n
          n.prev = t1.next
        t1 = t1.next

  def delete_start(self):
    if self.start is None:
      print("list is empty")
    else:
      self.start = self.start.next
      self.start.prev = None


  def delete_last(self):
    temp = self.start
    if temp is None:
      print("list is empty")
    else:
      while temp.next.next is not None:
        temp = temp.next
      temp.next.prev = None
      temp.next = None


  def delete_item(self,data):
    if self.start is None:
      print("list is empty")
    elif self.start.item == data:
      self.start = self.start.next
    else:
      temp =self.start
      if temp.item == data:
        temp = temp.start
      else:
        while temp.next is not None:
          if temp.next.item == data:
            temp.next.prev = temp
            temp.next = temp.next.next
            break
          temp =temp.next
        else:
          print("deleted value doesn't exist in the list")

  def __iter__(self):
    return Dll_iterator(self.start)




```
