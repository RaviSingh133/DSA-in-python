# Circular Doubly Link List

```python
class Node:

  def __init__(self,prev= None,item= None,next= None):
    self.prev = prev
    self.item = item
    self.next = next
```

```python
class Cdll:

  def __init__(self,start= None):
    self.start = start



  def list_empty(self):
    if self.start == None:
      print("list is empty")
    else:
      print("list is not empty")


  def insert_first(self,data):
    n= Node(None,data,None)
    if self.start == None:
      n.next =n
      n.prev = n

    else:
      n.next = self.start
      n.prev = self.start.prev
      self.start.prev.next = n
      self.start.prev = n
    self.start = n



  def print_list(self):
    if self.start == None:
      return self.list_empty()
    else:
      temp = self.start
      while True:
        print(temp.item,end = " ")
        temp = temp.next
        if temp == self.start:
          break
      print()



  def search_item(self,data):
    temp = self.start
    if temp.item == data:
      print(temp.item)
    else:
      temp = temp.next
      while temp != self.start:
        if temp.item == data:
          print(temp.item)
          break
        temp = temp.next
      else:
        print("searching value doesn't find")




  def insert_last(self,data):
    n = Node(None,data,None)
    if self.start == None:
      n.next = n
      n.prev = n
      self.start = n
    else:
      self.start.prev.next = n
      n.prev = self.start.prev
      self.start.prev = n
      n.next = self.start



  def insert_after(self,temp,data):
    n = Node(None,data,None)
    curr = self.start
    samp = self.start
    if self.start == None:
      return self.list_empty()
    else:
      while curr.next != samp:
        if curr.item == temp:
          n.next = curr.next
          n.prev = curr
          curr.next.prev = n
          curr.next = n
          break
        curr = curr.next
      else:
        if curr.item == temp:
          n.next = curr.next
          n.prev = curr
          curr.next.prev = n
          curr.next = n
        else:
          print("afer vlaue doesn't exist")



  def delete_first(self):
    if self.start == None:
      return self.list_empty()
    else:
      self.start.prev.next = self.start.next
      self.start.next.prev = self.start.prev
      self.start = self.start.next


  def delete_last(self):
    if self.start == None:
      return self.list_empty()
    else:
      self.start.prev.prev.next = self.start
      self.start.prev = self.start.prev.prev


  def delete_data(self,data):
    temp = self.start
    curr = self.start
    if self.start == None:
      return self.list_empty()
    else:
      if temp.item == data:
        return self.delete_first()
      else:
        while temp.next != curr:
          if temp.item == data:
            temp.prev.next = temp.next
            temp.next.prev = temp.prev
            break
          temp = temp.next
        if temp.next == curr:
          if temp.item == data:
            self.delete_last()
          else:
            print("delete value doesn't exist int the list")



  def __iter__(self):
    if self.start == None:
      return Cdll_iterator(None)
    else:
      return Cdll_iterator(self.start)


class Cdll_iterator:
  def __init__(self,start):
    self.current = start
    self.start = start
    self.count = 0
  def __iter__(self):
    return self
  def __next__(self):
    if self.current is None:
      raise StopIteration
    if self.current == self.start:
       if self.count == 1:
        raise StopIteration
    else:
      self.count = 1
    data = self.current.item
    self.current = self.current.next

    return data

```

```python
c1 = Cdll()
c1.list_empty()
c1.insert_first(20)
c1.insert_first(10)
c1.insert_first(30)
c1.insert_first(40)
c1.print_list()
c1.insert_last(5)
c1.insert_last(15)
c1.insert_last(80)
c1.print_list()
c1.insert_after(40,35)
c1.insert_after(80,62)
c1.insert_after(11,87)
c1.print_list()
print()
c1.delete_first()
c1.delete_first()
c1.print_list()
c1.delete_last()
c1.print_list()
c1.delete_data(30)
c1.delete_data(5)
c1.delete_data(80)
c1.delete_data(344)
c1.print_list()
```
