# Stack using List


```python
class Stack:
  def __init__(self):
    self.items = []

  def is_empty(self):
    if len(self.items) == 0:
      print("Stack is empty")
    else:
      print("Stack is not empty")

  def push(self,data):
    self.items.append(data)

  def pop(self):
    if len(self.items) == 0:
      raise IndexError("Stack is empty")
    else:
      self.items.pop()

  def peek(self):
    if len(self.items) == 0:
      raise IndexError("Stack is empty")
    else:
      print(self.items[-1])


  def lenght_stack(self):
    print(len(self.items))
```

```python
s1 = Stack()
s1.push(100)
s1.push(90)
s1.push(80)
s1.push(70)
s1.push(60)
s1.pop()
s1.pop()
s1.peek()
s1.lenght_stack()
```

Another approach

```python
class Stack1(list):
  def is_empty(self):
    if self == None:
      print("Stack is empty")
    else:
      print("Stack is not empty")


  def push(self,data):
    self.append(data)

  def pop(self):
    if self == None:
      raise IndexError("Stacl  is empty")
    else:
      return super().pop()

  def peek(self):
    if self == None:
      raise IndexError("Stack is empty")
    else:
      print("last value is ",self[-1])


  def lenght_stack(self):
    print("length of stack",len(self))


  def insert(self):
    raise AttributeError("Attribute Error 'Insert' doesn't valid in stack")



```

```python
s2 = Stack1()
s2.push(10)
s2.push(20)
s2.push(30)
s2.push(40)
s2.peek()
print("pop value is ",s2.pop())

```
