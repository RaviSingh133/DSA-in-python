# Stack using LinkList Concept

```python
class Node:
  def __init__(self,item=None,next=None):
    self.item = item
    self.next = next
```

```python
class Stack:
  def __init__(self,start= None):
    self.start = start
    self.count_item = 0

  def is_empty(self):
    if self.start == None:
      print("Stack is Empty")
    else:
      print("Stack is not Empty")

  def push(self,data):
    n = Node(data)
    n.next = self.start
    self.start = n
    self.count_item += 1

  def pop(self):
    if self.start == None:
      return self.is_empty()
    else:
      print("the pop value is ",self.start.item)
      self.start = self.start.next
      self.count_item -= 1


  def peek(self):
    if self.start == None:
      return self.is_empty()
    else:
      print("the last value is ",self.start.item)

  def total_item(self):
    print("total items is ",self.count_item)
```

```python
s1 = Stack()
s1.is_empty()
s1.push(100)
s1.push(90)
s1.push(80)
s1.total_item()
s1.pop()
s1.peek()
s1.total_item()
```
